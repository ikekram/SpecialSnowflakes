```python
lat_mid = 41.7867838
long_mid = -87.7550455
lat_ohare = 41.9837402
long_ohare = -87.977292
```

```python
import numpy as np

train['vert_dist_mid'] = lat_mid - train['Latitude']
train['horz_dist_mid'] = long_mid - train['Longitude']
train['mid'] = np.sqrt((train['vert_dist_mid']**2)+(train['horz_dist_mid']**2))


train['vert_dist_ohare'] = lat_ohare - train['Latitude']
train['horz_dist_ohare'] = long_ohare - train['Longitude']
train['ohare'] = np.sqrt((train['vert_dist_ohare']**2)+(train['horz_dist_ohare']**2))
train.head(5)
```

```python
train.ix[train['mid'] > train['ohare'], 'closer'] = 1 #map 1 to closer if Ohare is closer
train.ix[train['mid'] < train['ohare'], 'closer'] = 2 #map 2 to closer if mid is closer
```

```python
train.closer = train.closer.astype(int)
```
